```python
# Imports & helpers

import os, sys, json, glob, logging, warnings, math, hashlib, time
from contextlib import contextmanager
from types import SimpleNamespace
from pathlib import Path
from typing import Iterable, List, Dict, Tuple, Optional
import re

import numpy as np
import pandas as pd
from dateutil import tz
import matplotlib.pyplot as plt

from IPython.display import display, Image

class OneLineFormatter(logging.Formatter):
    def format(self, record):
        base = f"[{self.formatTime(record, '%Y-%m-%d %H:%M:%S')}] [{record.levelname}] {record.getMessage()}"
        if record.exc_info:
            base += " | " + self.formatException(record.exc_info).replace("\n", " | ")
        return base

def get_logger(name="START", level=logging.ERROR) -> logging.Logger:
    lg = logging.getLogger(name)
    lg.setLevel(level)
    handler = logging.StreamHandler(sys.stderr)
    handler.setLevel(level)
    handler.setFormatter(OneLineFormatter())
    lg.handlers.clear()
    lg.addHandler(handler)
    return lg

logger = get_logger()


# Paths & dirs

BASE_DIR      = Path.cwd()
DATA_DIR      = BASE_DIR / "data"
ARTIFACTS_DIR = BASE_DIR / "artifacts"

DIRS = SimpleNamespace(
    RAW       = DATA_DIR / "raw_data",
    PARQUET   = DATA_DIR / "parquet",
    MANIFEST  = DATA_DIR / "parquet" / "manifest",
    CLEANED   = DATA_DIR / "cleaned",
    QA        = DATA_DIR / "qa_reports",
    FIGURES   = ARTIFACTS_DIR / "figures",
    TABLES    = ARTIFACTS_DIR / "tables",
    REPORTS   = ARTIFACTS_DIR / "reports",
)

def ensure_dirs(d: SimpleNamespace = DIRS) -> None:
    """Create all required directories."""
    for p in [d.PARQUET, d.MANIFEST, d.CLEANED, d.QA, d.FIGURES, d.TABLES, d.REPORTS]:
        p.mkdir(parents=True, exist_ok=True)

ensure_dirs()
if not DIRS.RAW.exists():
    logger.error(f"Missing raw data folder: {DIRS.RAW}")


# Display / warnings
warnings.filterwarnings("once")
pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 200)


# Timezones & session helpers

ET  = tz.gettz("America/New_York")
UTC = tz.UTC

def to_utc(ts: pd.Series) -> pd.Series:
    """
    Convert timestamps to UTC.
    - If tz-aware: convert to UTC
    - If naive: assume ET, then convert to UTC
    """
    dt = pd.to_datetime(ts, errors="coerce", utc=False)
    if getattr(dt.dt, "tz", None) is not None:
        return dt.dt.tz_convert(UTC)
    return dt.dt.tz_localize(ET, nonexistent="shift_forward", ambiguous="NaT").dt.tz_convert(UTC)

def to_et(ts_utc: pd.Series) -> pd.Series:
    """UTC -> ET (tz-aware Series)."""
    return pd.to_datetime(ts_utc, utc=True).dt.tz_convert(ET)

def expected_bars_per_day(freq_min: int, open_str: str = "09:30", close_str: str = "16:00") -> int:
    t_open  = pd.to_datetime(open_str).time()
    t_close = pd.to_datetime(close_str).time()
    minutes = (pd.Timestamp.combine(pd.Timestamp.today(), t_close)
               - pd.Timestamp.combine(pd.Timestamp.today(), t_open)).seconds // 60
    return minutes // freq_min      # 78 for 5-min bars

def in_regular_session_mask(
    utc_ts: pd.Series,
    open_str: str = "09:30",
    close_str: str = "16:00",
    timestamp_is_bar_start: bool = True,
) -> pd.Series:
    """
    Boolean mask for UTC timestamps that fall within ET regular session.
    If timestamps mark bar *starts*: include [open, close).
    If timestamps mark bar *ends*:   include (open, close].
    """
    ts_et = pd.to_datetime(utc_ts, utc=True).dt.tz_convert(ET)
    hhmm  = ts_et.dt.strftime("%H:%M")
    if timestamp_is_bar_start:
        return (hhmm >= open_str) & (hhmm <  close_str)
    else:
        return (hhmm >  open_str) & (hhmm <= close_str)

def minute_of_day_et(ts_utc: pd.Series) -> pd.Series:
    """Minute-of-day in ET for a UTC timestamp series."""
    ts = to_et(ts_utc)
    return ts.dt.hour * 60 + ts.dt.minute


# DF utilities

def ensure_columns(df: pd.DataFrame, cols: Iterable[str], fill=np.nan) -> pd.DataFrame:
    """Add any missing columns with a default fill value; return same df for chaining."""
    for c in cols:
        if c not in df.columns:
            df[c] = fill
    return df

def coerce_numeric(df: pd.DataFrame, cols: Iterable[str]) -> pd.DataFrame:
    """In-place numeric coercion with NaN on errors; return df for chaining."""
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def safe_concat(frames: List[pd.DataFrame]) -> pd.DataFrame:
    frames = [f for f in frames if f is not None and not f.empty]
    if not frames:
        return pd.DataFrame()
    return pd.concat(frames, ignore_index=True)

def unique_path(path: Path) -> Path:
    """If path exists, append _1, _2, ... before the suffix to avoid overwriting."""
    if not path.exists():
        return path
    stem, suf = path.stem, path.suffix
    parent = path.parent
    k = 1
    while True:
        cand = parent / f"{stem}_{k}{suf}"
        if not cand.exists():
            return cand
        k += 1


# Save/load helpers

def save_json(obj, path: Path, indent: int = 2, overwrite: bool = True) -> Path:
    path.parent.mkdir(parents=True, exist_ok=True)
    if not overwrite:
        path = unique_path(path)
    path.write_text(json.dumps(json.loads(json.dumps(obj, default=str)), indent=indent))
    return path

def save_table(df: pd.DataFrame, name: str, dir_path: Path = DIRS.TABLES, fmt: str = "csv",
               index: bool = False, overwrite: bool = True) -> Path:
    """
    Save a table under artifacts/tables. fmt: 'csv' or 'parquet'.
    """
    dir_path.mkdir(parents=True, exist_ok=True)
    if fmt not in {"csv", "parquet"}:
        raise ValueError("fmt must be 'csv' or 'parquet'")
    path = dir_path / f"{name}.{fmt}"
    if not overwrite:
        path = unique_path(path)
    if fmt == "csv":
        df.to_csv(path, index=index)
    else:
        df.to_parquet(path, index=index, engine="pyarrow", compression="snappy")
    return path

def save_df_parquet(df: pd.DataFrame, path: Path, compression: str = "snappy", overwrite: bool = True) -> Path:
    """Write a single parquet file."""
    path.parent.mkdir(parents=True, exist_ok=True)
    if not overwrite:
        path = unique_path(path)
    df.to_parquet(path, index=False, engine="pyarrow", compression=compression)
    return path

def write_partitioned_parquet(
    df: pd.DataFrame,
    root: Path,
    by: List[str],
    filename_prefix: str = "part",
    compression: str = "snappy",
) -> List[Dict]:
    """
    Group by `by` (e.g., ['year','month']) and write one parquet per group.
    Outputs stable file names without full DataFrame sorts.
    """
    recs = []
    if not by:
        out = root / f"{filename_prefix}.parquet"
        save_df_parquet(df, out, compression=compression)
        recs.append({**{k: None for k in by}, "path": str(out), "rows": int(len(df))})
        return recs

    for keys, g in df.groupby(by, sort=True):
        keys = (keys,) if not isinstance(keys, tuple) else keys
        subdir = root
        name_parts = []
        for k, v in zip(by, keys):
            if isinstance(v, (int, np.integer)):
                subdir = subdir / f"{k}={int(v):02d}"
                name_parts.append(f"{int(v):02d}")
            else:
                subdir = subdir / f"{k}={v}"
                name_parts.append(str(v))
        subdir.mkdir(parents=True, exist_ok=True)
        out = subdir / f"{filename_prefix}_{'-'.join(name_parts)}.parquet"
        save_df_parquet(g.reset_index(drop=True), out, compression=compression)
        recs.append({**{k: (int(v) if isinstance(v, (int, np.integer)) else v) for k, v in zip(by, keys)},
                     "path": str(out), "rows": int(len(g))})
    return recs

def append_manifest(records: List[Dict], manifest_path: Path = DIRS.MANIFEST / "dataset_manifest.json") -> Path:
    """Append records (dicts) to a JSON manifest safely."""
    prev = []
    if manifest_path.exists():
        try:
            prev = json.loads(manifest_path.read_text())
        except Exception:
            prev = []
    return save_json(prev + records, manifest_path)

def save_fig(fig, name: str, dir_path: Path = DIRS.FIGURES, dpi: int = 150,
             show: bool = False, overwrite: bool = True) -> Path:
    dir_path.mkdir(parents=True, exist_ok=True)
    if not os.path.splitext(str(name))[1]:
        name = f"{name}.png"
    path = dir_path / name
    if not overwrite:
        path = unique_path(path)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")
    if show:
        plt.show()
    plt.close(fig)
    return path

def save_and_preview_table(
    df: pd.DataFrame,
    name: str,
    dir_path: Path = DIRS.TABLES,
    fmt: str = "csv",
    rows: int = 12,
    index: bool = False,
):
    p = save_table(df, name=name, dir_path=dir_path, fmt=fmt, index=index)
    print(f"Saved table → {p}")
    try:
        preview = pd.read_csv(p) if fmt == "csv" else pd.read_parquet(p)
        display(preview.head(rows))
    except Exception as e:
        print("Preview failed:", e)
    return p

def show_saved_figs(*names, dir_path: Path = DIRS.FIGURES, max_to_show: int = 12):
    exts = (".png", ".jpg", ".jpeg", ".svg", ".pdf")
    paths = []
    if names:
        for n in names:
            paths.append(dir_path / n if isinstance(n, str) else Path(n))
    else:
        paths = sorted([p for p in dir_path.iterdir() if p.suffix.lower() in exts])[:max_to_show]

    if not paths:
        print(f"No figures found in {dir_path}")
        return []

    shown = []
    print("Figures:")
    for p in paths:
        if Path(p).exists():
            print("  -", p)
            display(Image(filename=str(p)))
            shown.append(p)
        else:
            print("  - Missing:", p)
    return shown


# Repro / plotting

def set_seed(seed: int = 42):
    import random
    random.seed(seed); np.random.seed(seed)

def set_plot_defaults(figsize=(8, 5), dpi=120):
    plt.rcParams["figure.figsize"] = figsize
    plt.rcParams["figure.dpi"] = dpi
    plt.rcParams["axes.grid"] = False
    plt.rcParams["figure.autolayout"] = True


# Printing & timing

def print_section(title: str):
    bar = "─" * len(title)
    print(f"\n{title}\n{bar}")

@contextmanager
def timeblock(label: str):
    t0 = time.perf_counter()
    try:
        yield
    finally:
        dt = time.perf_counter() - t0
        print(f"⏱️ {label}: {dt:.2f}s")


# Integrity checks

def assert_bar_integrity(df: pd.DataFrame):
    """
    Basic structural checks before feature/label creation.
    """
    req = {"symbol","timestamp","open","high","low","close","volume"}
    missing = req - set(df.columns)
    assert not missing, f"Missing columns: {missing}"

    # duplicates
    dups = df.duplicated(["symbol","timestamp"]).sum()
    assert dups == 0, f"Duplicate (symbol,timestamp) rows: {dups}"

    # price bounds & non-negative volume
    bad_bounds = (df["low"] > df[["open","close","high"]].min(axis=1)) | \
                 (df["high"] < df[["open","close","low"]].max(axis=1)) | \
                 (df["volume"] < 0)
    assert not bad_bounds.any(), "Price/volume bounds violated"

    # 5-minute alignment check (best-effort)
    ts = pd.to_datetime(df["timestamp"], utc=True, errors="coerce")
    assert (ts.dt.minute % 5 == 0).all(), "Non 5-minute alignment detected"
```

```python
# Data Ingestion (For the sanity of this coursework, we are not including realtime data & api integration)

# Canonical schema for START
EXPECTED_COLS = [
    "timestamp","symbol","open","high","low","close","volume","vwap","trade_count"
]

# Column normalization for common provider names
def _normalize_start_cols(cols):
    alias = {
        "volume_weighted_price": "vwap",
        "vw": "vwap",
        "Symbol": "symbol",
    }
    out = []
    for c in cols:
        key = c if c in alias else c.lower()
        out.append(alias.get(key, c.lower()))
    return out

def _parse_symbol_from_filename(p: Path) -> str:
    m = re.match(r"([A-Za-z.\-]+)_", p.name)
    return m.group(1).upper() if m else p.stem.upper()

def _read_json_any(path: Path):
    """
    Read a JSON file that may be:
      - an array of objects
      - an object with wrapper key (data/bars/records/rows)
      - newline-delimited JSON (NDJSON)
    Returns a Python list of dicts.
    """
    try:
        text = path.read_text()
    except Exception as e:
        logger.error(f"Failed to read JSON: {path} | {e}", exc_info=True)
        raise

    data = None
    try:
        data = json.loads(text)
    except Exception:
        # try NDJSON
        try:
            data = [json.loads(line) for line in text.splitlines() if line.strip()]
        except Exception as e:
            logger.error(f"Failed to parse JSON: {path} | {e}", exc_info=True)
            raise

    # unwrap common container keys
    if isinstance(data, dict):
        for k in ("data","bars","records","rows"):
            if k in data and isinstance(data[k], list):
                data = data[k]
                break

    if isinstance(data, list):
        return data
    logger.error(f"Unsupported JSON shape in {path}: {type(data)}")
    return []

def _load_one_json_start(path: Path) -> pd.DataFrame:
    rows = _read_json_any(path)
    if not rows:
        return pd.DataFrame(columns=EXPECTED_COLS)

    df = pd.DataFrame(rows)
    # normalize columns & ensure presence
    df.columns = _normalize_start_cols(df.columns)
    if "symbol" not in df.columns:
        df["symbol"] = _parse_symbol_from_filename(path)
    else:
        df["symbol"] = df["symbol"].astype(str).str.upper().fillna(_parse_symbol_from_filename(path))
    if "vwap" not in df.columns: df["vwap"] = np.nan
    if "trade_count" not in df.columns: df["trade_count"] = np.nan
    for c in EXPECTED_COLS:
        if c not in df.columns:
            df[c] = np.nan
    df = df[EXPECTED_COLS]

    # numeric coercions (NaN on error)
    coerce_numeric(df, ["open","high","low","close","volume","vwap","trade_count"])

    # timestamp to UTC (assume ET if naive)
    df["timestamp"] = to_utc(df["timestamp"])
    df = df.dropna(subset=["timestamp"])

    # dedupe & deterministic order
    df = (df.drop_duplicates(subset=["symbol","timestamp"], keep="last")
            .sort_values(["timestamp","symbol"])
            .reset_index(drop=True))
    return df

def ingest_json_to_monthly_parquet_start(
    raw_dir: Path = DIRS.RAW,
    pq_root: Path = DIRS.PARQUET,
    manifest_path: Path = DIRS.MANIFEST / "dataset_manifest.json",
    compression: str = "snappy",
    verbose: bool = True,
    year_filter: Optional[List[int]] = [2024],
    session_only: bool = True,
    timestamp_is_bar_start: bool = True,
):
    files = sorted(glob.glob(str(raw_dir / "*.json")))
    if not files:
        logger.error(f"No JSON files under {raw_dir}")
        raise FileNotFoundError(f"No JSON files under {raw_dir}")

    parts = []
    for fp in files:
        dfp = _load_one_json_start(Path(fp))
        if not dfp.empty:
            parts.append(dfp)

    df_all = safe_concat(parts)
    if df_all.empty:
        logger.error("No rows parsed from JSON files.")
        raise RuntimeError("No rows parsed from JSON files.")

    # Session filter (ET)
    if session_only:
        mask = in_regular_session_mask(
            df_all["timestamp"], timestamp_is_bar_start=timestamp_is_bar_start
        )
        before = len(df_all)
        df_all = df_all[mask].copy()
        after = len(df_all)
        if verbose:
            print(f"Session filter applied: kept {after:,}/{before:,} rows (regular hours only)")

    # derive year/month in ET for partitioning
    ts_et = df_all["timestamp"].dt.tz_convert(ET)
    df_all["year"] = ts_et.dt.year.astype(int)
    df_all["month"] = ts_et.dt.month.astype(int)

    if year_filter:
        before = len(df_all)
        df_all = df_all[df_all["year"].isin(year_filter)].copy()
        after = len(df_all)
        if verbose:
            print(f"Year filter {year_filter}: kept {after:,}/{before:,} rows")

    # enforce canonical column order just before write
    df_all = df_all[EXPECTED_COLS + ["year","month"]]

    # write partitioned parquet under data/parquet/year=YYYY/month=MM/
    records = write_partitioned_parquet(
        df_all,
        root=pq_root,
        by=["year","month"],                 # creates .../year=YYYY/month=MM/
        filename_prefix="bars_5min",         # → bars_5min_YYYY-MM.parquet
        compression=compression
    )

    # append to manifest
    append_manifest(records, manifest_path=manifest_path)

    # concise artifacts: per-month counts
    month_tbl = (pd.DataFrame(records)
                 .astype({"year":"int64","month":"int64","rows":"int64"})
                 .sort_values(["year","month"]))
    save_table(month_tbl, name="ingestion_month_counts", dir_path=DIRS.TABLES, fmt="csv", index=False)

    if verbose:
        span = f"{df_all['timestamp'].min()} → {df_all['timestamp'].max()}"
        print(f"Ingestion complete | rows={len(df_all):,} | symbols={df_all['symbol'].nunique():,} | UTC span {span}")
        for r in records:
            print(f"  - {r['year']:04d}-{r['month']:02d}: {r['rows']:,} rows → {r['path']}")
        print(f"Manifest: {manifest_path}")
        print(f"Month summary: {DIRS.TABLES / 'ingestion_month_counts.csv'}")

    return df_all, month_tbl

print_section("Ingestion — Console Summary")

ing_tbl = DIRS.TABLES / "ingestion_month_counts.csv"
if ing_tbl.exists():
    print(f"Month counts table: { ing_tbl }")
    try:
        display(pd.read_csv(ing_tbl).head(12))
    except Exception as e:
        print("Preview failed:", e)
else:
    print("Missing:", ing_tbl)

manifest_path = DIRS.MANIFEST / "dataset_manifest.json"
if manifest_path.exists():
    try:
        manifest = json.loads(manifest_path.read_text())
        print(f"Manifest entries: {len(manifest)} | path: {manifest_path}")
    except Exception as e:
        print("Manifest read failed:", e)
else:
    print("Missing manifest:", manifest_path)

if 'df_ingested' in globals():
    top = df_ingested['symbol'].value_counts().head(10).rename_axis('symbol').reset_index(name='rows')
    save_and_preview_table(top, name="ingestion_top_symbols")
else:
    print("Note: df_ingested not in scope; skip top-symbols preview.")

# ---- Run ingestion
with timeblock("Ingestion"):
    df_ingested, month_summary = ingest_json_to_monthly_parquet_start(
        verbose=True,
        year_filter=[2024],
        session_only=True,          # regular session only (09:30–16:00 ET)
        timestamp_is_bar_start=True # typical for 5-min bars
    )
```

```python
# QA & CLEANING

NUM_COLS = ["open","high","low","close","volume","vwap","trade_count"]
SESSION_OPEN, SESSION_CLOSE = "09:30", "16:00"
FREQ_MIN = 5
EXP_BARS = expected_bars_per_day(FREQ_MIN, SESSION_OPEN, SESSION_CLOSE)  # 78

def _ohlc_bad(df: pd.DataFrame) -> pd.Series:
    """Flag rows that violate OHLC bounds; ignore rows with NaNs (not marked bad)."""
    o,h,l,c = df["open"], df["high"], df["low"], df["close"]
    cond = (
        (l.notna() & o.notna() & (l > o)) |
        (l.notna() & c.notna() & (l > c)) |
        (h.notna() & o.notna() & (h < o)) |
        (h.notna() & c.notna() & (h < c)) |
        (l.notna() & h.notna() & (l > h))
    )
    return cond.fillna(False)

def _vwap_outside(df: pd.DataFrame) -> pd.Series:
    return df["vwap"].notna() & ((df["vwap"] < df["low"]) | (df["vwap"] > df["high"]))

def _returns_and_outliers(df: pd.DataFrame, win: int = 50, z: float = 8.0) -> pd.DataFrame:
    """
    Add log return and robust outlier flags (ret + log-volume).
    Safe: treat non-positive closes as NaN for log.
    """
    d = df.sort_values(["symbol","timestamp"]).copy()

    # Safe log returns (ignore <= 0)
    close_safe = d["close"].where(d["close"] > 0)
    d["ret"] = np.log(close_safe).groupby(d["symbol"]).diff()

    # rolling median/MAD (robust) for returns
    med_ret = d.groupby("symbol")["ret"].transform(lambda s: s.rolling(win, min_periods=win//2).median())
    mad_ret = d.groupby("symbol")["ret"].transform(
        lambda s: s.rolling(win, min_periods=win//2).apply(lambda x: np.median(np.abs(x - np.median(x))), raw=True)
    )
    zret = (d["ret"] - med_ret) / (mad_ret.replace(0, np.nan) * 1.4826)
    d["outlier_ret"] = zret.abs() > z

    # robust log-volume outliers (ignore non-positive volumes)
    lv = np.log(d["volume"].where(d["volume"] > 0))
    med_lv = lv.groupby(d["symbol"]).transform(lambda s: s.rolling(win, min_periods=win//2).median())
    mad_lv = lv.groupby(d["symbol"]).transform(
        lambda s: s.rolling(win, min_periods=win//2).apply(lambda x: np.median(np.abs(x - np.median(x))), raw=True)
    )
    zv = (lv - med_lv) / (mad_lv.replace(0, np.nan) * 1.4826)
    d["outlier_vol"] = zv.abs() > z
    return d

def _symbol_day_coverage(df: pd.DataFrame) -> pd.DataFrame:
    ts_et = df["timestamp"].dt.tz_convert(ET)
    per_day = (
        pd.DataFrame({"symbol": df["symbol"].values,
                      "date_et": ts_et.dt.date.values,
                      "one": 1})
        .groupby(["symbol","date_et"])["one"].count()
        .rename("bars_found")
        .reset_index()
    )
    agg = per_day.groupby("symbol").agg(
        days=("date_et","nunique"),
        bars_found_total=("bars_found","sum")
    ).reset_index()
    agg["bars_expected_total"] = agg["days"] * EXP_BARS
    agg["pct_coverage"] = agg["bars_found_total"] / agg["bars_expected_total"]
    return agg

def _nan_counts_by_symbol(df: pd.DataFrame, cols=NUM_COLS) -> pd.DataFrame:
    nan_df = df[["symbol"] + cols].isna()
    sums = nan_df.groupby("symbol").sum().astype(int)
    sums.columns = [f"nan_{c}" for c in sums.columns]
    return sums.reset_index()

def qa_clean_one_month(dfm: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    """Return (cleaned_df, qa_table_per_symbol) for one month of long-format bars."""
    df = dfm.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"], utc=True)
    df["symbol"] = df["symbol"].astype(str)

    # Keep only regular session
    mask = in_regular_session_mask(df["timestamp"], SESSION_OPEN, SESSION_CLOSE)
    df = df.loc[mask].copy()

    # Drop exact duplicates on (symbol, timestamp)
    df = df.drop_duplicates(subset=["symbol","timestamp"], keep="last")

    # Sanity flags + safe fixes
    bad_vwap = _vwap_outside(df)
    bad_ohlc = _ohlc_bad(df)
    neg_vol  = df["volume"] < 0

    df.loc[bad_vwap, "vwap"] = np.nan
    df.loc[neg_vol,  "volume"] = np.nan

    # Add returns/outlier flags
    df = _returns_and_outliers(df)

    # QA aggregates per symbol
    cov   = _symbol_day_coverage(df)
    nans  = _nan_counts_by_symbol(df)
    ohlc  = bad_ohlc.groupby(df["symbol"]).sum().reset_index(name="ohlc_violations")
    vwapc = bad_vwap.groupby(df["symbol"]).sum().reset_index(name="vwap_outside")
    negvc = neg_vol.groupby(df["symbol"]).sum().reset_index(name="neg_volume_rows")
    outc  = df.groupby("symbol").agg(
        outlier_ret_count=("outlier_ret","sum"),
        outlier_vol_count=("outlier_vol","sum"),
        rows=("timestamp","count")
    ).reset_index()

    qa = (
        cov.merge(outc, on="symbol", how="left")
           .merge(ohlc, on="symbol", how="left")
           .merge(vwapc, on="symbol", how="left")
           .merge(negvc, on="symbol", how="left")
           .merge(nans, on="symbol", how="left")
           .sort_values("rows", ascending=False)
           .reset_index(drop=True)
    )

    cleaned = df.sort_values(["timestamp","symbol"]).reset_index(drop=True)
    cleaned = cleaned[["timestamp","symbol","open","high","low","close","volume","vwap","trade_count","ret","outlier_ret","outlier_vol"]]
    return cleaned, qa

def _parse_partition_int(name: str) -> int:
    """
    Accepts '2024' or 'year=2024' and returns 2024.
    Accepts '01'   or 'month=01' and returns 1.
    """
    if "=" in name:
        _, v = name.split("=", 1)
    else:
        v = name
    return int(v, 10)

def run_qa_cleaning_start(
    pq_root: Path = DIRS.PARQUET,
    cleaned_root: Path = DIRS.CLEANED,
    qa_out_csv: Path = DIRS.TABLES / "qa_summary.csv",
    cleaned_counts_csv: Path = DIRS.TABLES / "cleaned_month_counts.csv",
    verbose: bool = True,
):
    month_files = sorted(pq_root.glob("**/bars_5min_*.parquet"))
    if not month_files:
        raise FileNotFoundError(f"No monthly parquet under {pq_root}")

    qa_all, counts = [], []

    for p in month_files:
        # Support both 'year=YYYY/month=MM' and 'YYYY/MM'
        year_name  = p.parent.parent.name  # 'year=2024' or '2024'
        month_name = p.parent.name         # 'month=01' or '01'
        y = _parse_partition_int(year_name)
        m = _parse_partition_int(month_name)

        dfm = pd.read_parquet(p)
        cleaned, qa = qa_clean_one_month(dfm)

        out_dir  = cleaned_root / f"{y:04d}" / f"{m:02d}"
        out_file = out_dir / f"bars_5min_{y:04d}-{m:02d}_clean.parquet"
        save_df_parquet(cleaned, out_file)

        qa["year"], qa["month"] = y, m
        qa_all.append(qa)
        counts.append({"year": y, "month": m, "rows": int(len(cleaned))})

        if verbose:
            print(f"  - {y:04d}-{m:02d}: {len(cleaned):,} rows")

    qa_summary = pd.concat(qa_all, ignore_index=True)

    _ = save_table(qa_summary, name=qa_out_csv.stem, dir_path=qa_out_csv.parent, fmt="csv", index=False)
    month_counts = pd.DataFrame(counts).sort_values(["year","month"])
    _ = save_table(month_counts, name=cleaned_counts_csv.stem, dir_path=cleaned_counts_csv.parent, fmt="csv", index=False)

    if verbose:
        print("QA & Cleaning complete")
    return qa_summary, month_counts

def load_all_cleaned(cleaned_root=DIRS.CLEANED) -> pd.DataFrame:
    parts = []
    for p in sorted(cleaned_root.glob("*/**/*_clean.parquet")):
        dfm = pd.read_parquet(p)
        dfm["timestamp"] = pd.to_datetime(dfm["timestamp"], utc=True)
        dfm["symbol"] = dfm["symbol"].astype(str)
        cols = [c for c in ["timestamp","symbol",*NUM_COLS,"ret","outlier_ret","outlier_vol"] if c in dfm.columns]
        parts.append(dfm[cols])
    df = safe_concat(parts)
    if df.empty:
        raise RuntimeError("No cleaned data found.")
    return df.sort_values(["timestamp","symbol"]).reset_index(drop=True)

def full_year_missingness(low_coverage_threshold: float = 0.95):
    df_all = load_all_cleaned(DIRS.CLEANED)
    ts_et = df_all["timestamp"].dt.tz_convert(ET)
    df_all["date_et"] = ts_et.dt.date
    df_all["year"] = ts_et.dt.year
    df_all["month"] = ts_et.dt.month

    all_trading_dates = pd.Index(sorted(df_all["date_et"].unique()))
    n_days_year = len(all_trading_dates)

    by_symbol_rows = df_all.groupby("symbol")["timestamp"].size().rename("rows")
    by_symbol_nan  = df_all.groupby("symbol")[NUM_COLS].apply(lambda g: g.isna().sum()).astype(int)
    by_symbol_pct  = by_symbol_nan.div(by_symbol_rows, axis=0).add_suffix("_pct")

    symbol_tbl = pd.concat([by_symbol_rows, by_symbol_nan, by_symbol_pct], axis=1).reset_index()
    symbol_tbl["bars_expected_total"] = n_days_year * EXP_BARS
    symbol_tbl["coverage_pct"] = symbol_tbl["rows"] / symbol_tbl["bars_expected_total"]
    symbol_tbl["vwap_availability_pct"] = 1.0 - symbol_tbl["vwap_pct"].fillna(0.0)
    symbol_tbl_sorted = symbol_tbl.sort_values(["coverage_pct","rows"], ascending=[True, False])

    month_grp = df_all.groupby(["year","month"])
    month_rows = month_grp["timestamp"].size().rename("rows")
    month_nan  = month_grp[NUM_COLS].apply(lambda g: g.isna().sum()).astype(int)
    month_pct  = (month_nan.div(month_rows, axis=0).add_suffix("_pct")).reset_index()
    month_tbl  = (month_rows.reset_index()
                  .merge(month_nan.reset_index(), on=["year","month"])
                  .merge(month_pct, on=["year","month"])
                  .sort_values(["year","month"]))

    low_cov = symbol_tbl_sorted.loc[symbol_tbl_sorted["coverage_pct"] < low_coverage_threshold, 
                                    ["symbol","rows","coverage_pct"] + [f"{c}_pct" for c in NUM_COLS]]

    _ = save_table(symbol_tbl_sorted, name="full_year_by_symbol",     dir_path=DIRS.TABLES, fmt="csv", index=False)
    _ = save_table(month_tbl,        name="full_year_monthly_global", dir_path=DIRS.TABLES, fmt="csv", index=False)
    _ = save_table(low_cov,          name="low_coverage_symbols",     dir_path=DIRS.TABLES, fmt="csv", index=False)

    return symbol_tbl_sorted, month_tbl, low_cov

# ---------- Run QA & Cleaning ----------
with timeblock("QA & Cleaning"):
    qa_summary, cleaned_counts = run_qa_cleaning_start(verbose=True)

print_section("QA — Inline Preview")
display(qa_summary.head(12))
display(cleaned_counts)

with timeblock("Full-year missingness tables"):
    sym_full, month_full, low_cov = full_year_missingness(low_coverage_threshold=0.95)

print_section("Missingness — Inline Preview")
display(sym_full.head(12))
display(month_full.head(12))
display(low_cov.head(12))

# ---------- Visuals (inline) ----------
# 1) Coverage histogram
try:
    cov = sym_full["coverage_pct"].dropna()
    fig = plt.figure()
    plt.hist(cov, bins=40)
    plt.title("Coverage % Across Symbols (Full Year)")
    plt.xlabel("Coverage (0–1)"); plt.ylabel("Frequency")
    plt.show()
except Exception as e:
    print("Coverage histogram skipped:", e)

# 2) Monthly missingness (%) for key columns
try:
    cols_to_plot = ["open_pct","close_pct","volume_pct","vwap_pct","trade_count_pct"]
    mt = month_full[["year","month"] + cols_to_plot].copy()
    mt["yyyymm"] = mt["year"].astype(str) + "-" + mt["month"].astype(str).str.zfill(2)

    for col in cols_to_plot:
        fig = plt.figure()
        plt.plot(mt["yyyymm"], mt[col])
        plt.title(f"Monthly Missingness: {col.replace('_pct','').upper()}")
        plt.xlabel("Month"); plt.ylabel("Missing %")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
except Exception as e:
    print("Monthly missingness plots skipped:", e)

# 3) Outlier rate histograms
try:
    # outlier rates by symbol (share of rows flagged)
    rates = qa_summary[["symbol","outlier_ret_count","outlier_vol_count","rows"]].dropna()
    rates["ret_rate"] = rates["outlier_ret_count"] / rates["rows"]
    rates["vol_rate"] = rates["outlier_vol_count"] / rates["rows"]

    for col, title in [("ret_rate","Return Outlier Rate by Symbol"),
                       ("vol_rate","Volume Outlier Rate by Symbol")]:
        fig = plt.figure()
        plt.hist(rates[col].dropna(), bins=40)
        plt.title(title)
        plt.xlabel("Rate (0–1)"); plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()
except Exception as e:
    print("Outlier histograms skipped:", e)
```

```python
# EDA - windowed

# ---- CONFIG ----
DATE_START_ET = "2024-01-01"
DATE_END_ET   = "2024-03-31"   # ~3 months window for visuals
PREFERRED_SYMBOLS = ["SPY", "QQQ", "AAPL", "NVDA"]

def _parse_partition_int(name: str) -> int:
    # Works for '2024' or 'year=2024', '01' or 'month=01'
    if "=" in name:
        _, v = name.split("=", 1)
    else:
        v = name
    return int(v, 10)

def load_cleaned_window(cleaned_root=DIRS.CLEANED, start_et=DATE_START_ET, end_et=DATE_END_ET):
    start_et = pd.to_datetime(start_et).tz_localize(ET)
    end_et   = pd.to_datetime(end_et).tz_localize(ET)

    def ym_in_range(y, m):
        first = pd.Timestamp(year=y, month=m, day=1, tz=ET)
        last  = first + pd.offsets.MonthEnd(0)
        return not (last < start_et or first > end_et)

    parts = []
    for p in sorted((cleaned_root).glob("**/*_clean.parquet")):
        # Support 'YYYY/MM/file.parquet' or 'year=YYYY/month=MM/file.parquet'
        year_name  = p.parent.parent.name
        month_name = p.parent.name
        y = _parse_partition_int(year_name) if not year_name.isdigit() else int(year_name)
        m = _parse_partition_int(month_name) if not month_name.isdigit() else int(month_name)

        if ym_in_range(y, m):
            dfm = pd.read_parquet(p)
            dfm["timestamp"] = pd.to_datetime(dfm["timestamp"], utc=True)
            dfm["symbol"] = dfm["symbol"].astype(str)
            parts.append(dfm)

    df = safe_concat(parts)
    if df.empty:
        raise RuntimeError("No cleaned months matched the requested EDA window.")

    ts_et = df["timestamp"].dt.tz_convert(ET)
    mask = (ts_et >= start_et) & (ts_et <= end_et)
    df = df.loc[mask].copy()

    df["pct_return"] = df.groupby("symbol")["close"].pct_change()

    return df.sort_values(["timestamp","symbol"]).reset_index(drop=True)

def select_plot_symbols(df, preferred=PREFERRED_SYMBOLS, k=4):
    have = set(df["symbol"].unique().tolist())
    pick = [s for s in preferred if s in have]
    if len(pick) < min(k, len(have)):
        counts = df["symbol"].value_counts().index.tolist()
        for s in counts:
            if s not in pick and len(pick) < k:
                pick.append(s)
    return pick[:k]

# ---- SUMMARY TABLES ----
def compute_symbol_summary(df):
    ts_et = df["timestamp"].dt.tz_convert(ET)
    w = df.copy()
    w["date_et"] = ts_et.dt.date
    w["minute_of_day"] = ts_et.dt.hour * 60 + ts_et.dt.minute

    bars_per_day = (
        w.groupby(["symbol","date_et"])["timestamp"]
         .count().rename("bars_found").reset_index()
    )

    sym_stats = (
        w.groupby("symbol")
         .agg(
             rows=("timestamp","count"),
             days=("date_et","nunique"),
             mean_ret=("ret","mean"),
             std_ret=("ret","std"),
             median_ret=("ret","median"),
             mean_pct=("pct_return","mean"),
             std_pct=("pct_return","std"),
             median_pct=("pct_return","median"),
             mean_vol=("volume","mean"),
             median_vol=("volume","median"),
             vwap_nan=("vwap",  lambda s: s.isna().sum()),
             trade_nan=("trade_count", lambda s: s.isna().sum()),
         ).reset_index()
    )

    expected = expected_bars_per_day(5, "09:30", "16:00")  # 78
    bars_sum = bars_per_day.groupby("symbol")["bars_found"].sum().rename("bars_found_total").reset_index()
    cov = sym_stats[["symbol","days"]].merge(bars_sum, on="symbol", how="left")
    cov["bars_expected_total"] = cov["days"] * expected
    cov["pct_coverage"] = cov["bars_found_total"] / cov["bars_expected_total"]

    outliers = w.groupby("symbol").agg(outlier_ret=("outlier_ret","sum"),
                                       outlier_vol=("outlier_vol","sum")).reset_index()

    out = (sym_stats
           .merge(cov[["symbol","bars_found_total","bars_expected_total","pct_coverage"]], on="symbol", how="left")
           .merge(outliers, on="symbol", how="left")
           .sort_values("rows", ascending=False))
    return out

def descriptive_stats(df):
    numeric = [c for c in ["open","high","low","close","volume","vwap","trade_count","pct_return","ret"] if c in df.columns]
    return df[numeric].describe().T.reset_index().rename(columns={"index":"feature"})

def missing_and_outliers(df):
    fields = [c for c in ["open","high","low","close","volume","vwap","trade_count"] if c in df.columns]
    miss_counts = df[fields].isna().sum().to_frame("missing_rows").reset_index().rename(columns={"index":"feature"})
    miss_counts["missing_pct"] = miss_counts["missing_rows"] / len(df)
    out_ret = int(df["outlier_ret"].sum()) if "outlier_ret" in df.columns else 0
    out_vol = int(df["outlier_vol"].sum()) if "outlier_vol" in df.columns else 0
    out_tbl = pd.DataFrame({"metric":["outlier_ret_rows","outlier_vol_rows","total_rows"],
                            "value":[out_ret, out_vol, int(len(df))]})
    return miss_counts, out_tbl

# ---- PLOTS ----
def plot_hist_returns(df, symbol):
    s = df.loc[df["symbol"] == symbol, "pct_return"].dropna()
    if s.empty: return None
    fig = plt.figure()
    plt.hist(s, bins=120)
    plt.title(f"{symbol} — Histogram of 5-min returns")
    plt.xlabel("Return"); plt.ylabel("Frequency")
    return fig

def plot_corr_heatmap(df, symbols):
    sub = df[df["symbol"].isin(symbols)].copy()
    pivot = sub.pivot(index="timestamp", columns="symbol", values="pct_return").dropna(how="any")
    if pivot.shape[1] < 2:
        return None
    corr = pivot.corr()
    fig = plt.figure()
    plt.imshow(corr, interpolation="nearest")
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
    plt.yticks(range(len(corr.index)), corr.index)
    plt.title("Return Correlation (5-min)")
    plt.colorbar(); plt.tight_layout()
    return fig

def plot_intraday_mean_return(df, symbol):
    ts = df[df["symbol"] == symbol].copy()
    if ts.empty: return None
    ts_et = ts["timestamp"].dt.tz_convert(ET)
    ts["minute_of_day"] = ts_et.dt.hour * 60 + ts_et.dt.minute
    season = ts.groupby("minute_of_day")["pct_return"].mean()
    fig = plt.figure()
    season.plot()
    plt.title(f"{symbol} — Mean 5-min return by minute of day (ET)")
    plt.xlabel("Minute of Day (ET)"); plt.ylabel("Mean Return")
    return fig

def plot_daily_coverage_heatmap(df, symbols):
    ts = df[df["symbol"].isin(symbols)].copy()
    if ts.empty: return None
    ts_et = ts["timestamp"].dt.tz_convert(ET)
    ts["date_et"] = ts_et.dt.date
    daily = ts.groupby(["symbol","date_et"])["timestamp"].count().rename("bars").reset_index()
    heat = daily.pivot(index="symbol", columns="date_et", values="bars").fillna(0)
    fig = plt.figure()
    plt.imshow(heat, aspect="auto", interpolation="nearest")
    plt.title("Daily 5-min bar coverage — selected symbols")
    plt.xlabel("ET Date"); plt.ylabel("Symbol")
    plt.colorbar(); plt.tight_layout()
    return fig

def plot_price_line(df, symbol):
    ts = df[df["symbol"] == symbol].sort_values("timestamp").copy()
    if ts.empty: return None
    fig = plt.figure()
    plt.plot(ts["timestamp"], ts["close"])
    plt.title(f"{symbol} — Close Price (EDA Window)")
    plt.xlabel("UTC Timestamp"); plt.ylabel("Close")
    plt.tight_layout()
    return fig
def plot_intraday_volume_profile(df, symbol):
    ts = df[df["symbol"] == symbol].copy()
    if ts.empty: return None
    ts_et = ts["timestamp"].dt.tz_convert(ET)
    ts["minute_of_day"] = ts_et.dt.hour * 60 + ts_et.dt.minute
    prof = ts.groupby("minute_of_day")["volume"].mean()
    fig = plt.figure()
    prof.plot()
    plt.title(f"{symbol} — Mean Volume by Minute (ET)")
    plt.xlabel("Minute of Day (ET)"); plt.ylabel("Mean Volume")
    return fig

def plot_vwap_minus_close_hist(df, symbol):
    ts = df[(df["symbol"] == symbol) & df["vwap"].notna()].copy()
    if ts.empty: return None
    diff = ts["vwap"] - ts["close"]
    fig = plt.figure()
    plt.hist(diff, bins=120)
    plt.title(f"{symbol} — VWAP − Close (distribution)")
    plt.xlabel("VWAP − Close"); plt.ylabel("Frequency")
    return fig

def plot_return_autocorr(df, symbol, max_lag: int = 20):
    s = df.loc[df["symbol"] == symbol, "pct_return"].dropna()
    r = s.values
    if len(r) < max_lag + 2:
        return None
    ac = []
    for lag in range(1, max_lag + 1):
        a, b = r[:-lag], r[lag:]
        if len(a) < 2: break
        ac.append(np.corrcoef(a, b)[0, 1])
    fig = plt.figure()
    plt.bar(range(1, 1 + len(ac)), ac)
    plt.title(f"{symbol} — Autocorrelation of 5-min Returns")
    plt.xlabel("Lag"); plt.ylabel("Correlation")
    plt.tight_layout()
    return fig

def plot_multi_hist_returns(df, symbols):
    """Small grid of return histograms for 3–4 symbols (pairplot vibe)."""
    keep = [s for s in symbols if s in df["symbol"].unique()][:4]
    if not keep: return None
    n = len(keep)
    # one figure, multiple axes (still one chart per figure guideline = 1 fig, multiple subplots OK if needed)
    fig, axes = plt.subplots(1, n, figsize=(4*n, 3), dpi=120)
    if n == 1: axes = [axes]
    for ax, sym in zip(axes, keep):
        s = df.loc[df["symbol"] == sym, "pct_return"].dropna()
        ax.hist(s, bins=100)
        ax.set_title(sym)
    fig.suptitle("5-min Return Distributions (EDA Window)")
    fig.tight_layout()
    return fig

# ---- DRIVER ----
def run_eda_full(start_et=DATE_START_ET, end_et=DATE_END_ET, preferred=PREFERRED_SYMBOLS, save=False, show_inline=True):
    df = load_cleaned_window(DIRS.CLEANED, start_et, end_et)
    basic = {
        "rows": len(df),
        "symbols": df["symbol"].nunique(),
        "utc_start": str(df["timestamp"].min()),
        "utc_end": str(df["timestamp"].max()),
    }
    print(f"EDA load: rows={basic['rows']:,} | symbols={basic['symbols']:,} | UTC {basic['utc_start']} → {basic['utc_end']}")

    # Choose symbols for plots
    symbols = select_plot_symbols(df, preferred)
    print("Plot symbols:", symbols)
    sym0 = symbols[0]

    # Tables: symbol-level summary, descriptive stats, missingness + outliers
    summary = compute_symbol_summary(df)
    desc = descriptive_stats(df)
    miss_tbl, out_tbl = missing_and_outliers(df)

    display(summary.head(12))
    display(desc)
    display(miss_tbl)
    display(out_tbl)

    if save:
        _ = save_table(summary, "eda_symbol_summary", fmt="csv", index=False)
        _ = save_table(desc,    "eda_descriptive_stats", fmt="csv", index=False)
        _ = save_table(miss_tbl,"eda_missing_values",    fmt="csv", index=False)
        _ = save_table(out_tbl, "eda_outlier_counts",    fmt="csv", index=False)

    figs = [
        (plot_hist_returns(df, sym0),                 f"{sym0}_hist_returns.png"),
        (plot_corr_heatmap(df, symbols),             "correlation_heatmap.png"),
        (plot_intraday_mean_return(df, sym0),        f"{sym0}_intraday_mean_return.png"),
        (plot_daily_coverage_heatmap(df, symbols),   "coverage_heatmap.png"),
        (plot_price_line(df, sym0),                  f"{sym0}_price_line.png"),
        (plot_intraday_volume_profile(df, sym0),     f"{sym0}_intraday_volume.png"),
        (plot_vwap_minus_close_hist(df, sym0),       f"{sym0}_vwap_minus_close.png"),
        (plot_return_autocorr(df, sym0, 20),         f"{sym0}_autocorr.png"),
        (plot_multi_hist_returns(df, symbols),       "multi_hist_returns.png"),
    ]
    for fig, name in figs:
        if fig is None: continue
        if save:
            _ = save_fig(fig, name, DIRS.FIGURES, show=False)
        if show_inline:
            plt.show()

    return df, summary, symbols, basic

# ---- RUN ----
with timeblock("EDA — Complete"):
    eda_df, eda_summary, plot_symbols, eda_basic = run_eda_full(
        DATE_START_ET, DATE_END_ET, PREFERRED_SYMBOLS, save=False, show_inline=True
    )
```

```python
# DATA PREPROCESSING PIPELINE (train-aware, leak-safe)

# -------- Config (edit these dates as needed) --------
MODEL_START_ET = "2024-01-01"
MODEL_END_ET   = "2024-03-31"      # preprocessing window
TRAIN_END_ET   = "2024-03-15"      # split point: [start, TRAIN_END] -> train; (TRAIN_END, end] -> valid

FREQ_MIN = 5
OPEN_STR, CLOSE_STR = "09:30", "16:00"

# Paths
from pathlib import Path
FEAT_DIR = (Path.cwd() / "data" / "features"); FEAT_DIR.mkdir(parents=True, exist_ok=True)
SPEC_PATH = Path.cwd() / "artifacts" / "reports" / "preprocess_spec.json"; SPEC_PATH.parent.mkdir(parents=True, exist_ok=True)

import numpy as np
import pandas as pd

# -------- Utilities --------
def _mad(x: pd.Series) -> float:
    med = np.nanmedian(x)
    return np.nanmedian(np.abs(x - med))

def _parse_partition_int(name: str) -> int:
    """Accepts '2024' or 'year=2024'; '01' or 'month=01'."""
    if "=" in name:
        _, v = name.split("=", 1)
    else:
        v = name
    return int(v, 10)

def load_cleaned_window_for_model(cleaned_root=DIRS.CLEANED, start_et=MODEL_START_ET, end_et=MODEL_END_ET) -> pd.DataFrame:
    start_et = pd.to_datetime(start_et).tz_localize(ET)
    end_et   = pd.to_datetime(end_et).tz_localize(ET)

    def ym_in_range(y, m):
        first = pd.Timestamp(year=y, month=m, day=1, tz=ET)
        last  = first + pd.offsets.MonthEnd(0)
        return not (last < start_et or first > end_et)

    parts = []
    for p in sorted(Path(cleaned_root).glob("**/*_clean.parquet")):
        year_name  = p.parent.parent.name
        month_name = p.parent.name
        y = _parse_partition_int(year_name)  if not year_name.isdigit()  else int(year_name)
        m = _parse_partition_int(month_name) if not month_name.isdigit() else int(month_name)
        if ym_in_range(y, m):
            dfm = pd.read_parquet(p)
            dfm["timestamp"] = pd.to_datetime(dfm["timestamp"], utc=True)
            dfm["symbol"] = dfm["symbol"].astype(str)
            parts.append(dfm)

    df = pd.concat(parts, ignore_index=True) if parts else pd.DataFrame()
    if df.empty:
        raise RuntimeError("No cleaned months matched the requested preprocessing window.")
    ts_et = df["timestamp"].dt.tz_convert(ET)
    df = df.loc[(ts_et >= start_et) & (ts_et <= end_et)].copy()
    return df.sort_values(["timestamp","symbol"]).reset_index(drop=True)

def engineer_features(df: pd.DataFrame) -> pd.DataFrame:
    """Past-only features, per symbol."""
    d = df.sort_values(["symbol","timestamp"]).copy()

    # Ensure returns present
    if "ret" not in d.columns:
        d["ret"] = np.log(d["close"]).groupby(d["symbol"]).diff()
    d["pct_return"] = d.groupby("symbol")["close"].pct_change()

    g = d.groupby("symbol", sort=False)

    # Lags
    for k in range(1, 6):
        d[f"ret_lag{k}"] = g["ret"].shift(k)

    # Rolling stats
    d["range_pct"]    = (d["high"] - d["low"]) / d["close"]
    d["roll_vol20"]   = g["ret"].transform(lambda s: s.rolling(20, min_periods=10).std())
    d["roll_range20"] = g["range_pct"].transform(lambda s: s.rolling(20, min_periods=10).mean())

    # Volume surprise (robust z): guard against zeros/NaNs in early windows
    vol_med = g["volume"].transform(lambda s: s.rolling(50, min_periods=25).median())
    vol_mad = g["volume"].transform(
        lambda s: s.rolling(50, min_periods=25)
                  .apply(lambda x: np.median(np.abs(x - np.median(x))), raw=True)
    )
    lv = np.log(d["volume"].where(d["volume"] > 0))
    lv_med = np.log(vol_med.where(vol_med > 0))
    denom = (vol_mad * 1.4826).replace(0, np.nan)
    d["vol_surp"] = (lv - lv_med) / denom

    # VWAP – Close & missing flags
    d["vwap_close"] = d["vwap"] - d["close"]
    d["isna_vwap"]  = d["vwap"].isna().astype(int)
    d["isna_trade"] = d["trade_count"].isna().astype(int)

    # Time features
    ts = d["timestamp"].dt.tz_convert(ET)
    minute = ts.dt.hour * 60 + ts.dt.minute
    d["minute_sin"] = np.sin(2*np.pi*minute/(60*24))
    d["minute_cos"] = np.cos(2*np.pi*minute/(60*24))
    d["dow"]        = ts.dt.dayofweek.astype(int)

    # Cross-sectional "market" return and its lag
    mkt = d.groupby("timestamp")["ret"].median().rename("mkt_ret")
    d = d.merge(mkt, left_on="timestamp", right_index=True, how="left")
    d["mkt_ret_lag1"] = d.groupby("symbol", sort=False)["mkt_ret"].shift(1)

    # Target: next-bar up/down (classification)
    d["y_up"] = (g["close"].shift(-1) > d["close"]).astype("float32")

    return d

def split_train_valid(df, train_end_et=TRAIN_END_ET):
    te = pd.to_datetime(train_end_et).tz_localize(ET)
    ts = df["timestamp"].dt.tz_convert(ET)
    train = df.loc[ts <= te].copy()
    valid = df.loc[ts >  te].copy()
    return train, valid

def fit_preprocessor(train: pd.DataFrame, num_cols: list, clip_cfg: dict):
    """Compute medians for imputation, robust clip stats, and standardization (fit on train)."""
    spec = {"median": {}, "clip": {}, "scale": {}, "ohe": {}}

    # Imputation medians (feature-wise)
    for c in num_cols:
        spec["median"][c] = float(np.nanmedian(train[c].values))

    # Winsorization stats (median, MAD) per feature
    for c, k in clip_cfg.items():
        med = float(np.nanmedian(train[c].values))
        mad = float(_mad(train[c].values))
        spec["clip"][c] = {"median": med, "mad": mad, "k": float(k)}

    # Standardization stats
    for c in num_cols:
        mu = float(np.nanmean(train[c].values))
        sd = float(np.nanstd (train[c].values, ddof=0))
        spec["scale"][c] = {"mean": mu, "std": sd if sd > 0 else 1.0}

    # OHE categories for DOW (freeze from training)
    spec["ohe"]["dow"] = sorted([int(v) for v in pd.Series(train["dow"]).dropna().unique().tolist()])

    return spec

def apply_preprocessor(df: pd.DataFrame, spec: dict, num_cols: list):
    X = df.copy()

    # Impute vwap_close with 0 to preserve "distance meaning" (flag already present)
    if "vwap_close" in X.columns:
        X["vwap_close"] = X["vwap_close"].fillna(0.0)

    # Generic median imputation
    for c, med in spec["median"].items():
        if c in X.columns:
            X[c] = X[c].fillna(med)

    # Robust winsorization
    for c, s in spec["clip"].items():
        if c in X.columns:
            med, mad, k = s["median"], s["mad"], s["k"]
            if not (mad is None or mad == 0):
                lo = med - k * 1.4826 * mad
                hi = med + k * 1.4826 * mad
                X[c] = X[c].clip(lower=lo, upper=hi)

    # Standardize
    for c, sc in spec["scale"].items():
        if c in X.columns:
            mu, sd = sc["mean"], sc["std"] if sc["std"] != 0 else 1.0
            X[c] = (X[c] - mu) / sd

    # OHE for DOW with frozen categories
    cats = spec["ohe"]["dow"]
    X["dow"] = pd.Categorical(X["dow"], categories=cats)
    X = pd.get_dummies(X, columns=["dow"], prefix="dow", dummy_na=False)

    return X

def assemble_feature_matrix(df_eng: pd.DataFrame, spec=None):
    """Create X/y/meta, apply spec if given; otherwise return raw engineered + configs."""
    # Candidate numeric features
    base_num = [
        "ret_lag1","ret_lag2","ret_lag3","ret_lag4","ret_lag5",
        "roll_vol20","range_pct","roll_range20","vol_surp",
        "vwap_close","minute_sin","minute_cos","mkt_ret_lag1"
    ]
    # Include contemporaneous 'ret' to predict next bar
    base_num.append("ret")

    # Indicator flags
    ind_cols = ["isna_vwap","isna_trade"]

    keep_cols = ["timestamp","symbol","close","y_up"] + base_num + ind_cols + ["dow"]
    Z = df_eng[keep_cols].copy()

    # Clipping config (robust)
    clip_cfg = {
        "ret": 8.0, "ret_lag1": 8.0, "ret_lag2": 8.0, "ret_lag3": 8.0, "ret_lag4": 8.0, "ret_lag5": 8.0,
        "vol_surp": 6.0, "range_pct": 6.0, "roll_range20": 6.0, "roll_vol20": 6.0, "vwap_close": 6.0,
        "mkt_ret_lag1": 8.0
    }
    num_cols = base_num  # standardized/imputed set

    if spec is None:
        # Fitting pass
        return Z, num_cols, clip_cfg
    else:
        X = apply_preprocessor(Z, spec, num_cols)
        feature_cols = [c for c in X.columns if c not in ("timestamp","symbol","close","y_up")]
        return X[feature_cols], Z["y_up"].astype("float32"), Z[["timestamp","symbol","close"]], feature_cols

# -------- Run preprocessing end-to-end --------
with timeblock("Preprocessing"):
    # 1) Load window & engineer features
    df_all = load_cleaned_window_for_model()
    df_eng = engineer_features(df_all)

    # 2) Drop rows with NA target (last bar per symbol)
    df_eng = df_eng[df_eng["y_up"].notna()].reset_index(drop=True)

    # 3) Split train/valid by time
    train_df, valid_df = split_train_valid(df_eng)

    # 4) Fit spec on training
    Ztrain, num_cols, clip_cfg = assemble_feature_matrix(train_df, spec=None)
    spec = fit_preprocessor(Ztrain, num_cols=num_cols, clip_cfg=clip_cfg)
    save_json(spec, SPEC_PATH)  # silent save

    # 5) Apply spec to train/valid
    X_train, y_train, meta_train, feat_list = assemble_feature_matrix(train_df, spec=spec)
    X_valid, y_valid, meta_valid, _        = assemble_feature_matrix(valid_df, spec=spec)

    # 6) Save datasets & artifacts
    train_out = pd.concat([meta_train, y_train.rename("y"), X_train], axis=1)
    valid_out = pd.concat([meta_valid, y_valid.rename("y"), X_valid], axis=1)
    (FEAT_DIR / "train.parquet").parent.mkdir(parents=True, exist_ok=True)
    train_out.to_parquet(FEAT_DIR / "train.parquet", index=False)
    valid_out.to_parquet(FEAT_DIR / "valid.parquet", index=False)
    save_table(pd.DataFrame({"feature": feat_list}), name="feature_list", dir_path=DIRS.TABLES, fmt="csv", index=False)

    # Quick summary (inline only)
    summary = pd.DataFrame(
        {"split":["train","valid"],
         "rows":[len(train_out), len(valid_out)],
         "features":[len(feat_list), len(feat_list)],
         "symbols":[meta_train["symbol"].nunique(), meta_valid["symbol"].nunique()]}
    )
    display(summary)

    # Base rate on train (useful for classifier calibration)
    base_rate = float(y_train.mean())
    print(f"y base rate (train): {base_rate:.3f}")

    # Preview (inline)
    display(train_out.head(8))
    display(valid_out.head(8))
```
